Prediccion del Potencial de un Jugador

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('data_fifa.csv')
df.head(4)
```

```python
df.describe()
```

```python
print(df['Age'].isnull().any(), df['Potential'].isnull().any())
```

```python
plt.scatter(data = df, x = 'Age', y = 'Potential')
plt.xlabel('Age')
plt.ylabel('Potential');
```

```python
plt.hist2d(data = df, x = 'Age', y = 'Potential')
plt.xlabel('Age')
plt.ylabel('Potential');
plt.colorbar()
```

```python
# Data traning: x_train = Age, y_train = Potential
# Data testing: x_test = Age, y_test = Potential
```

Scikit_learn

```python
from sklearn.model_selection import train_test_split
```

```python
# dividir
x_train, x_test, y_train,  y_test = train_test_split(df['Age'], df['Potential'], test_size = 0.2)
# print(x_train, x_test, y_train,  y_test)
print(x_train.shape, y_train.shape)
print(type(x_train))
```

```python
# Convert to ndarray
x_train = np.array(x_train)
x_test = np.array(x_test)

y_train = np.array(y_train)


# Reshage
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)

y_train = y_train.reshape(-1, 1)
```

```python
# Seleccionar el modelo
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
# Entrenamiento
regression.fit(x_train, y_train)
```

```python
# Prediction
y_predict = regression.predict(x_test)
```

```python
print(x_test)
print('\n')
print(y_predict)
```

```python
# encontrar el error de prediccion
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_predict)
```
